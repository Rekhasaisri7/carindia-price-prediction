```python
a['model_name1'] = a['model_name'].str.extract(r'(\S{3,}|\S{1,2}\s+\S+)',expand =True)
```

```python

a['model_name1'].unique()
```

```python
a
```

```python

a['model_name1'].value_counts()
```

```python
b = a.drop(['maker','city','fuel_type','model_name'],axis = 1)
b
```

```python
c = b.loc[(b.model_name1 == 'Wagon')]
c
```

```python
d = b.loc[(b.model_name1 == 'Swift')]
d
```

```python
e = b.loc[(b.model_name1 == 'i10')]
e
```

```python
f = pd.concat([c,d,e])
f
```

```python
g = pd.get_dummies(f,dtype = int)
g
```

```python
y = g['price (₹)']
y
```

```python
x = g.drop('price (₹)',axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=42)
x_train.head(10)
```

```python
y_test.tail(10)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
ypred = reg.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ypred,y_test)
```

```python
res = pd.DataFrame(columns=['price','predicted'])
res['price'] = y_test
res['predicted'] = ypred
res = res.reset_index()
res['ID'] = res.index
res.head(10)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='ID',y='price',data=res.head(50))
sns.lineplot(x='ID',y='predicted',data=res.head(50))
plt.plot()
```
